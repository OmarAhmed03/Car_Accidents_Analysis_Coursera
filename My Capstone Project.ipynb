## Importing Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing The dataset

```python
df_acc = pd.read_csv('https://opendatatc.blob.core.windows.net/opendatatc/NCDB_2017.csv')
```

```python
df_acc.head()
```

## Cleaning the dataset

```python
nan_value = float("NaN")
df_acc.replace("UU", nan_value, inplace=True)
df_acc.replace("U", nan_value, inplace=True)
df_acc.replace("UUUU", nan_value, inplace=True)
df_acc.replace("X", nan_value, inplace=True)
df_acc.replace("XX", nan_value, inplace=True)
df_acc.replace("XXXX", nan_value, inplace=True)
df_acc.replace("Q", nan_value, inplace=True)
df_acc.replace("QQ", nan_value, inplace=True)
df_acc.replace("QQQQ", nan_value, inplace=True)
df_acc.replace("N", nan_value, inplace=True)
df_acc.replace("NN", nan_value, inplace=True)
df_acc.replace("NNNN", nan_value, inplace=True)
df_acc.dropna(inplace=True)
```

```python
df_acc.drop(['V_ID', 'P_ID','C_CASE','C_WDAY', 'C_TRAF', 'C_VEHS','P_PSN', 'C_YEAR'], axis=1, inplace = True)
```

```python
df_acc.reset_index(drop=True, inplace=True)
df_acc
```

## Number of accidents every Month

```python
mnth = df_acc['C_MNTH'].value_counts()
mnth = mnth.to_frame()
mnth.index.rename('Month', inplace=True)
mnth = mnth.rename(columns={'C_MNTH':'Total'})
mnth.reset_index(inplace=True)
mnth = mnth.astype({"Month": int})
mnth.sort_values(by="Month", inplace=True)
```

```python
mnth.plot(kind='bar', x='Month', figsize=(10, 6))
plt.title('Accidents according to months')
plt.show()
```

## Number of accidents every hour

```python
hour = df_acc['C_HOUR'].value_counts()
hour = hour.to_frame()
hour.index.rename('Hour', inplace=True)
hour = hour.rename(columns={'C_HOUR':'Total'})
hour.reset_index(inplace=True)
hour = hour.astype({"Hour": int})
hour.sort_values(by="Hour", inplace=True)
hour
```

```python
hour.plot(kind='bar', x='Hour', figsize=(10, 6))
plt.title('Accidents according to Hour')
plt.show()
```

## Relationship between weather and accidents

```python
weather = df_acc['C_WTHR'].value_counts()
weather = weather.to_frame()
weather.index.rename('weather', inplace=True)
weather = weather.rename(columns={'C_WTHR':'Total'})
weather
```

```python
weather.plot(kind='bar')
plt.title('Accidents according to Weather')
positions = (0, 1, 2,3,4,5,6)
labels = ("Clear and sunny", "Raining", "Overcast", "Snowing", "Freezing", "xFreezing", "Strong wind")
plt.xticks(positions, labels)
plt.show()
```

## Relationship between age and accidents

```python
age = df_acc['P_AGE'].value_counts()
age = age.to_frame()
age.index.rename('age', inplace=True)
age = age.rename(columns={'P_AGE':'Total'})
age.reset_index(inplace=True)
age = age.astype({"age": int})
age.sort_values(by="age", inplace=True)
age
```
